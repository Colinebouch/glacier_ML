# Clean the data set for modeling

## Add WH, WHsin, and dummy variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_main = pd.read_csv('/home/colili/Documents/PhD/project_john/DataFrame_produced_clean/df_slope_point.csv', index_col=0)
df_main.drop(columns=['index'], inplace=True)
```

```python
df_main['WHsin'] = df_main['thickness']*df_main['width_centerline']*np.sin(df_main['Bed_slope'])
```

```python
df_main['WH']=df_main['thickness']/df_main['width_centerline']
```

```python
df_main['dummy']=np.random.rand(df_main.shape[0])
```

```python
df_main.columns
```

## Recalculate SMB with beta, H and ELA

```python
df_main['SMB_calc'] = df_main['beta']*(df_main['surface_elevation']-df_main['ELA'])
```

## Drop glaciers shorter than km

```python
df_main = df_main[df_main.Lmax > 1000].copy()
```

## Drop glaciers that already melt away (A<km2)

```python
df_main = df_main[df_main.Area > 1].copy()
```

## Drop surge index=9 (not present in Svalbard)

```python
df_main.drop(df_main[df_main['Surge']==9].index, inplace=True)
```

## Convert surge index

```python
df_main['Surge'] = df_main['Surge'].replace({3: 1, 2:1})
```
