```python
xdata = df[['Area', 'Zmin', 'Zmax', 'Zmed', 'Lmax', 'thick_w0',
       'thick_w1', 'thick_w2', 'thick_w3', 'thick_w4', 'thick_w5', 'thick_w6',
       'thick_w7', 'thick_w8', 'thick_w9', 'thick_w10', 'thick_w11',
       'windowlength_w0', 'windowlength_w1', 'windowlength_w2',
       'windowlength_w3', 'windowlength_w4', 'windowlength_w5',
       'windowlength_w6', 'windowlength_w7', 'windowlength_w8',
       'windowlength_w9', 'windowlength_w10', 'windowlength_w11',
       'bedslope_w0', 'bedslope_w1', 'bedslope_w2', 'bedslope_w3',
       'bedslope_w4', 'bedslope_w5', 'bedslope_w6', 'bedslope_w7',
       'bedslope_w8', 'bedslope_w9', 'bedslope_w10', 'bedslope_w11',
       'surfslope_w0', 'surfslope_w1', 'surfslope_w2', 'surfslope_w3',
       'surfslope_w4', 'surfslope_w5', 'surfslope_w6', 'surfslope_w7',
       'surfslope_w8', 'surfslope_w9', 'surfslope_w10', 'surfslope_w11'
           ]]

ybasic = df[['surgetype-basic']]
yreduced = df[['surgetype-reduced']]
```

```python
KF = KFold(n_splits = 5, shuffle = True)
KF_index = KF.split(df)

predictions_basic = []
predictions_reduced = []
true = []
xgmodels = []
trains = []
tests = []

for train, test in KF_index:
    trains.append(train)
    tests.append(test)
    
    xtrain = xdata.iloc[train]
    xtest = xdata.iloc[test]
    
    ytrain = ybasic.iloc[train]
    ytest = ybasic.iloc[test]
    
    true.append(ytest)
    
    # basic
    dtrain = xgb.DMatrix(xtrain, label=ytrain)
    dtest = xgb.DMatrix(xtest, label=ytest)
    
    bst = xgb.train(param, dtrain, num_boost_round=20000,)
    xgmodels.append(bst)
    
    predictions_basic.append(bst.predict(dtest))
    
    # reduced
    ytest = ybasic.iloc[test]
    dtest = xgb.DMatrix(xtest, label=ytest)
    predictions_reduced.append(bst.predict(dtest))
    
```

```python
def get_mean_tpr(data, model):
    fprs = []
    tprs = []
    mean_fpr = np.linspace(0, 1, 100)
    for k in np.arange(0, 5, 1):
        fpr, tpr, threshold = metrics.roc_curve(
            y_true=data.iloc[tests[k]]
            , y_score=model[k]
        )
        tpr = np.interp(mean_fpr, fpr, tpr)
        fprs.append(fpr)
        tprs.append(tpr)
        print(data.iloc[tests[k]].sum())
        
    return mean_fpr, np.array(tprs).mean(axis=0)

basic_roc = get_mean_tpr(ybasic, predictions_basic)
reduced_roc = get_mean_tpr(yreduced, predictions_reduced)
```

```python
fig, ax = plt.subplots(figsize=(9, 8))

aucs = []
for k in np.arange(0, 5, 1):
    auc = metrics.roc_auc_score(y_true=ybasic.iloc[tests[k]], y_score=predictions_basic[k])
    aucs.append(auc)
    
basic_auc = np.mean(aucs).round(3)

aucs = []
for k in np.arange(0, 5, 1):
    auc = metrics.roc_auc_score(y_true=yreduced.iloc[tests[k]], y_score=predictions_reduced[k])
    aucs.append(auc)

reduced_auc = np.mean(aucs).round(3)

fpr, tpr = basic_roc
ax.plot(fpr, tpr, color='purple', label='Basic Classification AUC={}'.format(basic_auc), linewidth=2)

fpr, tpr = reduced_roc
ax.plot(fpr, tpr, color='green', label='Reduced Classification AUC={}'.format(reduced_auc), linewidth=2)
ax.plot([0, 1], [0, 1], color='black', alpha=0.5, linestyle='--')

ax.set_ylabel('True Positive Rate', fontsize=15)
ax.set_xlabel('False Positive Rate', fontsize=15)

ax.legend(fontsize=15)
```
